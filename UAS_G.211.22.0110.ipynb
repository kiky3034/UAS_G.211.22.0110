```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
dataset = pd.read_csv('data_hujan.csv')
print(dataset)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Membuat Variabel di masing-masing data dari dataset
Bulan = (dataset['Bulan'])
A =  (dataset['1981-1990'])
B =  (dataset['1991-2000'])
C =  (dataset['2001-2010'])


# style
sns.set(style="darkgrid")

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist([Bulan]*3, bins=12, weights=[A,B,C], color=["deepskyblue", "crimson", "darkgreen"], label=["1981-1990", "1991-2000", "2001-2010"])
plt.xlabel("Bulan")
plt.ylabel("Curah Hujan (mm)")
plt.xticks(dataset['Bulan'])
plt.yticks([0, 100, 200, 300, 400, 500, 600, 700])
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
dataset = pd.read_csv('data_suhu.csv')
print(dataset)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

Waktu = (dataset['Waktu'])
A =  (dataset['A'])
B =  (dataset['B'])
C =  (dataset['C'])

#Style
sns.set(style="whitegrid")

# Membuat multi line plot
plt.figure(figsize=(10, 5))
plt.plot(Waktu, A, marker='o', label='A', color="deepskyblue", linewidth=3)
plt.plot(Waktu, B, marker='o', label='B', color="lime", linewidth=3)
plt.plot(Waktu, C, marker='o', label='C', color="tomato", linewidth=3)

# Menambahkan label pada sumbu x dan y
plt.xlabel('Waktu')
plt.ylabel('Suhu (°C)')

# Menambahkan judul grafik
plt.title('Perubahan Suhu Permukaan')

# Memutar label tanggal dan jam agar tidak saling berdekatan/bertumpangan
plt.xticks(rotation=45)
plt.yticks([21, 22, 23, 24, 25, 26, 27])
plt.grid(False, axis='x')

# Menambahkan legenda
plt.legend()

# Menampilkan grafik
plt.show()

```
